```python
df = pd.read_csv("diamonds.csv")
```

```python
df.head()
```

```python
del df["Unnamed: 0"]
```

```python
df.head(5)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X = df["price"].values.reshape(-1, 1)
```

```python
y = df.drop(["price"], axis=1)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from math import sqrt
import numpy as np

print(mean_squared_error(y_test, y_pred))
print(sqrt(mean_squared_error(y_test, y_pred)))
print(r2_score(y_test, y_pred))
```

```python
parameters = {"alpha" : np.arange(0.1, 1.1, 0.1)}
```

```python
lasso = GridSearchCV(Lasso(), parameters).fit(X_train, y_train)
```

```python
y_l_pred = lasso.predict(X_test)
```

```python
print(mean_squared_error(y_test, y_l_pred))
print(sqrt(mean_squared_error(y_test, y_l_pred)))
print(r2_score(y_test, y_l_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures

p = PolynomialFeatures(4)
X_train_p = p.fit_transform(X_train)
X_test_p = p.fit_transform(X_test)
```

```python
lr = LinearRegression().fit(X_train_p, y_train)
```

```python
y_pred = lr.predict(X_test_p)
```

```python
print(mean_squared_error(y_test, y_pred))
print(sqrt(mean_squared_error(y_test, y_pred)))
print(r2_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor, export_text
from sklearn.metrics import classification_report, accuracy_score
from sklearn import tree
```

```python
tree_reg = DecisionTreeRegressor(max_depth=4).fit(X_train, y_train)
```

```python
y_pred = tree_reg.predict(X_test)
```

```python
print(mean_squared_error(y_test, y_pred))
print(sqrt(mean_squared_error(y_test, y_pred)))
print(r2_score(y_test, y_pred))
```

```python
tree_reg.get_params()
```

```python
param_grid={'max_depth': range(2, 10),
            'min_samples_split': range(2, 10)}
```

```python
tree_reg_b = GridSearchCV(tree_reg, param_grid , cv=5, n_jobs=-1 ,verbose=True).fit(X_train, y_train)
```

```python
y_pred = tree_reg_b.predict(X_test)
```

```python
print(mean_squared_error(y_test, y_pred))
print(sqrt(mean_squared_error(y_test, y_pred)))
print(r2_score(y_test, y_pred))
```

```python
import tensorflow as tf
import keras

y = df["price"]
X = df.drop(["price"], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
nn_reg = tf.keras.Sequential()
```

```python
nn_reg.add(tf.keras.layers.Dense(64, activation='relu', input_shape=(9,)))
nn_reg.add(tf.keras.layers.Dense(32, activation="relu"))
# Dropout позволяет внести фактор случайности - при обучении часть нейронов будет отключаться
# каждый нейрон, в данном случае, будет отключаться с вероятностью 0.1
nn_reg.add(tf.keras.layers.Dropout(0.1))
nn_reg.add(tf.keras.layers.Dense(16, activation="relu"))
nn_reg.add(tf.keras.layers.Dropout(0.1))
nn_reg.add(tf.keras.layers.Dense(1))
```

```python
nn_reg.summary()
```

```python
nn_reg.compile(optimizer=tf.keras.optimizers.Adam(learning_rate = 0.005), loss="mse")
```

```python
nn_reg.fit(X_train, y_train, epochs=20)
```

```python
y_pred = nn_reg.predict(X_test)
```

```python
print(mean_squared_error(y_test, y_pred))
print(sqrt(mean_squared_error(y_test, y_pred)))
print(r2_score(y_test, y_pred))
```

```python
df_c = pd.read_csv("weatherAUS.csv")
```

```python
df_c.head(5)
```

```python
df_c.info()
```

```python
df_c.describe()
```

```python
df_c = df_c.rename (
    columns = {
        'Date' : 'date',
        'Location' : 'location',
        'MinTemp' : 'min_temp',
        'MaxTemp' : 'max_temp',
        'Rainfall' : 'rainfall',
        'Evaporation' : 'evaporation',
        'Sunshine' : 'sunshine',
        'WindGustDir' : 'wind_gust_dir',
        'WindGustSpeed' : 'wind_gust_speed',
        'WindDir9am' : 'wind_dir_9am',
        'WindDir3pm' : 'wind_dir_3pm',
        'WindSpeed9am' : 'wind_speed_9am',
        'WindSpeed3pm' : 'wind_speed_3pm',
        'Humidity9am' : 'humidity_9am',
        'Humidity3pm' : 'humidity_3pm',
        'Pressure9am' : 'pressure_9am',
        'Pressure3pm' : 'pressure_3pm',
        'Cloud9am' : 'cloud_9am',
        'Cloud3pm' : 'cloud_3pm',
        'Temp9am' : 'temp_9am',
        'Temp3pm' : 'temp_3pm',
        'RainToday' : 'rain_today',
        'RainTomorrow' : 'rain_tomorrow'
    }
)
```

```python
df_c.head(5)
```

```python
df_c.isnull().sum()
```

```python
for column in df_c.columns:
    if df_c[column].dtype == 'int64':
        df_c[column].fillna(df_c[column].median(), inplace=True)
    elif df_c[column].dtype == 'float64':
        df_c[column].fillna(df_c[column].mean(), inplace=True)
    else:
        df_c[column].fillna(df_c[column].mode()[0], inplace=True)
```

```python
df_c.isnull().sum()
```

```python
df_c.duplicated().sum()
```

```python
df_c.info()
```

```python
df_class = df_c.drop(['date'], axis=1)
df_class['location'] = pd.Categorical(df_class['location']).codes
df_class['wind_gust_dir'] = pd.Categorical(df_class['wind_gust_dir']).codes
df_class['wind_dir_9am'] = pd.Categorical(df_class['wind_dir_9am']).codes
df_class['wind_dir_3pm'] = pd.Categorical(df_class['wind_dir_3pm']).codes
df_class['rain_today'] = pd.Categorical(df_class['rain_today']).codes
df_class['rain_tomorrow'] = pd.Categorical(df_class['rain_tomorrow']).codes
```

```python
df_class.head(10)
```

```python
for col in df_class.columns:
    df_class.hist(col)
```

```python
y = df_class["rain_tomorrow"]
X = df_class.drop(["rain_tomorrow"], axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

normalizer = StandardScaler()

X_normalized_zscore = normalizer.fit_transform(X)
```

```python
X_normalized_zscore[y==0].shape
```

```python
X_normalized_zscore[y==1].shape
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler()

X_resampled, y_resampled = rus.fit_resample(X_normalized_zscore, y)

X_resampled[y_resampled==0].shape
```

```python
X_resampled[y_resampled==1].shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn10 = KNeighborsClassifier(n_neighbors=10, p=2)
knn100 = KNeighborsClassifier(n_neighbors=100,  p=2)
```

```python
knn10.fit(X_train,y_train)
knn100.fit(X_train,y_train)
```

```python
print(accuracy_score(y_test, knn10.predict(X_test)))
print(accuracy_score(y_test, knn100.predict(X_test)))
```

```python
print(classification_report(y_test, knn10.predict(X_test)))
print(classification_report(y_test, knn100.predict(X_test)))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib as plt

print(roc_auc_score(y_test, knn10.predict(X_test)))
print(roc_auc_score(y_test, knn100.predict(X_test)))
```
