## 2.1 Понижение размерности

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier

import xgboost as xgb

from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import GridSearchCV

from sklearn.feature_selection import VarianceThreshold, SelectKBest

import joblib
```

```python
def print_metrics(y_pred, y_true):
  print(f'''
confusion_matrix: \n{confusion_matrix(y_pred, y_test)}

classification report: {classification_report(y_pred, y_test)}
  ''')
```

```python
X = data.drop(columns='Revenue')
y = data['Revenue']
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(X)
X, y = RandomOverSampler(random_state=42).fit_resample(X, y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, shuffle=True)
```

```python
y.value_counts()
```

исправили дисбаланс классов, чтобы оценки качества моделей стали более информативными (accuracy)


Поскольку наблюдений (строк) не такое уж большое кол-во, размер тестовой выборки с 0.3 сократим до 0.2. Также используем параметр shuffle для наболее объективных оценок на тестовой выборке

```python
gbc = GradientBoostingClassifier()


params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.001],
    'max_depth': [2,3,4,5]}
grid_search = GridSearchCV(gbc, param_grid=params, cv=5)

grid_search.fit(X_train, y_train)

```

```python
y_pred = grid_search.predict(X_test)
```

```python
print_metrics(y_pred, y_test)
```

```python
joblib.dump(gbc, "models/gradient_boosting_classifier_model")
```

```python
knn = KNeighborsClassifier()

params = {'n_neighbors': [1,2,3,4,5,6,7]}
grid = GridSearchCV(estimator=knn, param_grid=params, cv=5)
grid.fit(X_train, y_train)
```

```python
knn_best = grid.best_estimator_
y_pred = grid.predict(X_test)
print_metrics(y_pred, y_test)
```

```python
joblib.dump(knn_best, "models/knn_best_model")
```

```python
xgb_classifier = xgb.XGBClassifier()

params = {'learning_rate':[0.05, 0.001],
          'max_depth': [3,4,5]}
grid_search = GridSearchCV(xgb_classifier, param_grid=params, cv=5)
grid_search.fit(X_train, y_train)
```

```python
xgb_best = grid_search.best_estimator_
```

```python
print_metrics(grid_search.predict(X_test), y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()

gnb.fit(X_train, y_train)
print_metrics(gnb.predict(X_test), y_test)
```
