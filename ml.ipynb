```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("weatherAUS.csv")
```

```python
data.isna().sum()
```

```python
for col in data.columns:
    if data[col].dtype == "int64" :
        data[col].fillna(data[col].mean(), inplace=True)
    elif data[col].dtype == "float64":
        data[col].fillna(data[col].median(), inplace=True)
    else:
        data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates()
```

```python
data.info()
```

```python
categorical = ["Location", "WindGustDir", "WindDir9am", "WindDir3pm", "RainToday", "RainTomorrow"]

for categor in categorical:
    data[categor] = pd.Categorical(data[categor]).codes
```

```python
data
```

```python
data["Date"] = pd.to_datetime(data["Date"])
data["Date_Day"] = data["Date"].dt.day
data["Date_Month"] = data["Date"].dt.month
data["Date_Year"] = data["Date"].dt.year

data = data.drop("Date", axis=1)

data
```

```python
for col in data.columns:
    sns.boxplot(data[col])
    plt.show()
```

```python
for col in data.columns:
    data.hist(col)
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

data_filtred = data[~((data < Q1 - IQR*1.5)|(data > Q3 + IQR*1.5)).any(axis=1)]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm

from sklearn.ensemble import GradientBoostingClassifier
from catboost import CatBoostClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
y = data["RainTomorrow"]

X = data.drop("RainTomorrow", axis=1)

print(y)
print(X)
```

```python
from sklearn.preprocessing import StandardScaler
from imblearn.under_sampling import RandomUnderSampler

X_skaler = StandardScaler().fit_transform(X)
X_skaler
```

```python
y.value_counts()
```

```python
X_sampler, y_sampler = RandomUnderSampler().fit_resample(X_skaler, y)

y_sampler.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sampler, y_sampler, test_size=0.2)
```

```python
models = {}

def print_accuracy_model(model, name):
    models[name] = model
    y_pred = model.predict(X_test)
    print(accuracy_score(y_pred, y_test))
    print(classification_report(y_pred, y_test))
```

# KNN

```python
knn = KNeighborsClassifier()

params = {"n_neighbors": range(5, 50)}

grid_knn = GridSearchCV(estimator=knn, param_grid=params, cv=5)
grid_knn.fit(X_train, y_train)
```

```python
knn_best = grid_knn.best_estimator_
print_accuracy_model(grid_knn, "KNN")
```

# GaussianNB

```python
gnb = GaussianNB()

gnb.fit(X_test, y_test)
```

```python
print_accuracy_model(gnb, "GNB")
```

# SVM

```python
svm = svm.SVC()

svm.fit(X_train, y_train)
```

```python
print_accuracy_model(svm, "SVM")
```

# GradientBoosting

```python
gradboosting = GradientBoostingClassifier()

gradboosting.fit(X_train, y_train)
```

```python
print_accuracy_model(gradboosting, "GradBoosting")
```

# CatBoosting


```python
catboosting = CatBoostClassifier()

catboosting.fit(X_train, y_train)
```

```python
print_accuracy_model(catboosting, "CatBoosting")
```

# DesiciontTree

```python
tree = DecisionTreeClassifier()

tree.fit(X_train, y_train)
```

```python
print_accuracy_model(tree, "DesTree")
```
